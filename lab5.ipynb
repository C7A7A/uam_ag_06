```python
from qiskit import QuantumCircuit
from qiskit import Aer, transpile, assemble
from numpy import pi
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
circ = QuantumCircuit(6, 3)

circ.h(0)
circ.h(1)
circ.h(2)
circ.barrier()

circ.x(0)
circ.x(2)
circ.ccx(0, 2, 5)
circ.x(0)
circ.x(2)
circ.barrier()

circ.cx(2, 4)
circ.barrier()

circ.x(0)
circ.cx(2, 3)
circ.cx(0, 3)
circ.x(0)
circ.barrier()

circ.h(0)
circ.h(1)
circ.h(2)
circ.barrier()

circ.measure(0, 0)
circ.measure(1, 1)
circ.measure(2, 2)

circ.draw(output="mpl")
```

```python
simulator = Aer.get_backend('qasm_simulator')
telas = assemble(circ, shots=100_000)
results = simulator.run(telas)

hist = results.result().get_counts()
print(hist)
```
