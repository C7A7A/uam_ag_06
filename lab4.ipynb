```python
from qiskit import QuantumCircuit
from qiskit import Aer, transpile, assemble
from numpy import pi
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
circ = QuantumCircuit(3)

circ.x(0)
circ.x(1)
circ.barrier()

circ.h(2)
circ.cp(pi/2, 1, 2)
circ.cp(pi/4, 0, 2)
circ.h(1)
circ.cp(pi/2, 0, 1)
circ.h(0)

circ.swap(0, 2)

circ.measure_all()
circ.draw(output="mpl")
```

```python
simulator = Aer.get_backend('qasm_simulator')
telas = assemble(circ, shots=100_000)
results = simulator.run(telas)

hist = results.result().get_counts()
print(hist)
```

```python
circ = QuantumCircuit(4, 3)

circ.h(0)
circ.h(1)
circ.h(2)
circ.x(3)
circ.barrier()

circ.cp(pi/4, 0, 3)
circ.cp(pi/4, 0, 3)
circ.cp(pi/4, 0, 3)
circ.cp(pi/4, 0, 3)

circ.cp(pi/4, 1, 3)
circ.cp(pi/4, 1, 3)

circ.cp(pi/4, 2, 3)

circ.barrier()

circ.h(0)
circ.cp(-pi/2, 0, 1)

circ.h(1)
circ.cp(-pi/4, 0, 2)
circ.cp(-pi/2, 1, 2)

circ.h(2)
circ.barrier()

circ.measure(0, 0)
circ.measure(1, 1)
circ.measure(2, 2)
circ.draw(output="mpl")
```

```python
simulator = Aer.get_backend('qasm_simulator')
telas = assemble(circ, shots=100_000)
results = simulator.run(telas)

hist = results.result().get_counts()
print(hist)
```
